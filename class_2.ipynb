### Adding integers

Define two variables that hold integers, add them together, and print the result. 

Your code should also print subtraction of these numbers, division, and exponent i.e. x^y.

````python
var1 = int(10)
var2 = int(20)
var3 = var1 + var2
var4 = var1 - var2
var5 = var1 / var2
var6 = var1**var2

print (int(var3))
print (int(var4))
print (int(var5))
print (int(var6))

# OR...in a shorter version
print ("------------------")

var01 = int(10)
var02 = int(20)

print (int(var01 + var02))
print (int(var01 - var02))
print (int(var01 / var02))
print (int(var01**var02))
````

### Using operators

Simplify the code below using the `+=` symbol:
```python
number1 = int(input("First number: "))
number2 = int(input("Second number: "))
number3 = int(input("Third number: "))

sum = number1 + number2 + number3
print("The sum of the numbers: ", sum)
```

````python
total = 0

number1 = int(input("First number: "))
number2 = int(input("Second number: "))
number3 = int(input("Third number: "))

total += number1
total += number2
total += number3

print("The sum of the numbers: ", total)
````
